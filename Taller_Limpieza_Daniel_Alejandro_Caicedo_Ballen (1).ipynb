# Ejercicio 9: Groupby

Agrupar los datos por categorías, nos permite realizar análisis más profundos y obtener insights valiosos, que a simple vista no vemos en la base de datos.

1. Crea un nuevo dataframe llamado "df_sexo" donde se agrupe por 'sexo' y se calcule la suma de las columnas 'poblacion_total' y 'ocupados'. Crea una nueva columna que se llame 'porcentaje_ocupados' donde dividas las columna 'ocupados' sobre 'poblacion_total'. Al final usa reset_index() para convertir las categorías de sexo en una columna.

```python
df_sexo = df_empleo.groupby("sexo")[["poblacion_total", "ocupados"]].sum()
```

```python
df_sexo["porcentaje_ocupados"] = df_sexo["ocupados"] / df_sexo["poblacion_total"]
```

```python
df_sexo = df_sexo.reset_index()
```

2. Crea un nuevo dataframe llamado "df_educacion" donde se agrupe por 'nivel_educativo' y se calcule la suma de las columnas 'poblacion_total' y 'ocupados'. Crea una nueva columna que se llame 'porcentaje_ocupados' donde dividas las columna 'ocupados' sobre 'poblacion_total'. Al final usa reset_index() para convertir las categorías de nivel educativo en una columna.

```python
df_educacion = df_empleo.groupby("nivel_educativo")[["poblacion_total", "ocupados"]].sum()
```

```python
df_educacion["porcentaje_ocupados"] = df_educacion["ocupados"] / df_educacion["poblacion_total"]
```

```python
df_educacion = df_educacion.reset_index()
```
