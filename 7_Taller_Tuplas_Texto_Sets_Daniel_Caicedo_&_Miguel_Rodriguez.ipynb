### Ejercicio 25: Pipeline de Limpieza (List -> Set -> List)
**Contexto:** Tienes una lista de precios con duplicados y desordenada. Quieres una lista limpia y ordenada.

```python
precios_raw = [100, 200, 100, 50, 200]
```

```python
precios_raw=set(precios_raw)
precios_limpios=sorted(list(precios_raw))
print(precios_limpios)
```
