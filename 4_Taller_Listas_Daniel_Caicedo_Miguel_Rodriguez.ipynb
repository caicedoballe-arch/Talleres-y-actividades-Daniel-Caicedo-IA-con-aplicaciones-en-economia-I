## **Bloque 1: Gesti√≥n de √çndices e Indicadores**
Trabajaremos con la lista `ipc_mensual = [0.5, 0.8, 1.2, 0.4, 0.3, 0.9, 1.1]` que representa la inflaci√≥n mensual.

```python
ipc_mensual = [0.5, 0.8, 1.2, 0.4, 0.3, 0.9, 1.1]
```

```python
# 1. Agrega el dato de inflaci√≥n del nuevo mes: 0.6 al final.
ipc_mensual.append(0.6)
print(ipc_mensual)

```

```python
# 2. Inserta el valor 0.7 en la segunda posici√≥n (√≠ndice 1) por un ajuste retroactivo.
ipc_mensual.insert(1, 0.7)
ipc_mensual

```

```python
# 3. El valor 1.2 fue un error de digitaci√≥n. C√°mbialo por 1.0.
ipc_mensual=[0.5, 0.7, 0.8, 1.2, 0.4, 0.3, 0.9, 1.1, 0.6]
posicion=ipc_mensual.index(1.2)
ipc_mensual[posicion]=1.0
ipc_mensual

```

```python
# 4. Extrae los primeros 3 meses en una lista llamada 'primer_trimestre'.
primer_trimestre=[]
for mes in range (3):
    primer_trimestre.append(ipc_mensual[mes])
primer_trimestre
    

```

```python
# 5. Extrae los √∫ltimos 2 meses en una lista llamada 'cierre_a√±o'.
cierre_a√±o=[]
for mes in range (-2,0):
    cierre_a√±o.append(ipc_mensual[mes])
cierre_a√±o

```

```python
# 6. Ordena la lista de menor a mayor inflaci√≥n.
inflacion_ordenada=sorted(ipc_mensual)
print(inflacion_ordenada)


```

```python
# 7. Encuentra el √≠ndice donde qued√≥ el valor m√°ximo (1.1) y gu√°rdalo en 'pos_max'.
maximo = max(ipc_mensual)
pos_max = ipc_mensual.index(maximo)
print(pos_max)
```

```python
# 8. Elimina el valor m√°s bajo de la lista (0.3).
minimo=min(ipc_mensual)
ipc_mensual.remove(minimo)
ipc_mensual

```

```python
# 9. Cuenta cu√°ntas veces aparece el valor 0.6 y gu√°rdalo en 'conteo_06'.
conteo_06=0
for i in ipc_mensual:
    if i==0.6:
        conteo_06+=1
print(conteo_06)

```

```python
# 10. Invierte el orden actual de la lista.
lista_invertida=ipc_mensual[::-1]
lista_invertida

```

## **Bloque 2: Listas de Listas - Microdatos de Empresas**
Gestionar√°s la siguiente matriz `empresas = [['E1', 500, True], ['E2', 150, False], ['E3', 1200, True]]`.
Cada sublista corresponde a `[Nombre_Empresa, N√∫mero_Empleados, Cotiza_Bolsa]`.

```python
empresas = [['E1', 500, True], ['E2', 150, False], ['E3', 1200, True]]
```

```python
# 11. Accede al n√∫mero de empleados de 'E2' y gu√°rdalo en 'emp_e2'.
for empresa in empresas:
    if empresa[0] == 'E2':
        emp_e2 = empresa[1]
print(emp_e2)    
```

```python
# 12. Cambia el estado de 'E2' a True (ahora cotiza en bolsa).
for empresa in empresas:
    if empresa[0] == 'E2':
        empresa[2] = True
print(empresas)    
```

```python
# 13. Agrega una nueva empresa: ['E4', 800, False].
nueva_empresa=['E4', 800, False]
empresas.append(nueva_empresa)
empresas
```

```python
# 14. Calcula el total de empresas en la matriz y gu√°rdalo en 'n_empresas'.
n_empresas=0
for empresa in empresas:
    n_empresas+=1
print(n_empresas)

```

```python
# 15. Crea una lista vac√≠a 'nombres' y usa un 'for' para llenarla con los nombres (E1, E2...).
nombres=[]
for i in empresas:
    nombres.append(i[0])
nombres
```

```python
# 16. Calcula el promedio de empleados y gu√°rdalo en 'promedio_emp'.
contador=0
suma=0
for i in empresas:
    suma+=i[1]
    contador+=1
    promedio=suma/contador
promedio_emp=promedio
promedio_emp

```

```python
# 17. Elimina la sublista de la empresa 'E1'.
empresas.pop(0)
empresas
```

```python
# 18. Accede al nombre de la √∫ltima empresa de la lista y gu√°rdalo en 'ultima_emp'.
ultima_emp=empresas[-1][0]
ultima_emp
```

```python
# 19. Crea una lista 'grandes' con las empresas que tienen > 1000 empleados (trae la sublista con la informaci√≥n completa de la o las empresas).
grandes=[]
for empresa in empresas:
    if empresa[1]>1000:
        grandes.append(empresa)
grandes
```

```python
# 20. Modifica el nombre de 'E2' a 'Empresa Lider' dentro de la matriz.
empresas[0][0]="Empresa Lider"
empresas
```

```python
# --- VERIFICACI√ìN AUTOM√ÅTICA - NO MODIFICAR ESTA CELDA ---
assert emp_e2 == 150
assert empresas[0][0] == 'Empresa Lider'
assert n_empresas == 4
assert nombres == ['E1', 'E2', 'E3', 'E4']
assert promedio_emp == 662.5
assert len(grandes) == 1
print("¬°Taller finalizado exitosamente!")
```
